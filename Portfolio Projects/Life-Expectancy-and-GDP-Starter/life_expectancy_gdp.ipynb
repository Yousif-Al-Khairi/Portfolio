```python
#importing neccesary packages and loading dataset into dataframe 
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
df = pd.read_csv("all_data.csv")
```

```python
#Investigates each countrys relationship between its GDP and life expectancy
def generate_LE_from_GDP(country):
    filtered_df = df[df["Country"] == f"{country}"]
    sns.lmplot(x='GDP', y='LE', data=filtered_df, markers=['o'])
    plt.title(f"Graph of {country}'s GDP versus its life expectancy")
    plt.xlabel("GDP")
    plt.ylabel("Life expectancy in years")
    plt.show()
    plt.clf()

for country in df["Country"].unique():
    generate_LE_from_GDP(country)
```

```python
#Investigates each countrys relationship between its GDP over the years 2000 - 2015
def gen_graph(country):
    filtered_df = df[df["Country"] == f"{country}"]
    ax = plt.subplot()
    sns.barplot(x='Year', y='GDP', data=filtered_df)
    plt.title(f"Graph of {country}'s GDP  over the years 2000-2015")
    ax.set_xticks(ax.get_xticks())  # Get the current tick positions
    ax.set_xticklabels(df['Year'].unique(), rotation=45, ha='right')
    plt.xlabel("Year")
    plt.ylabel("GDP")
    plt.xticks()
    plt.show()
    plt.clf()

for country in df["Country"].unique():
    generate_LE_from_GDP(country)
```

```python
def graph(country):
    filtered_df = df[df["Country"] == f"{country}"]
    ax = plt.subplot()
    sns.barplot(x='Year', y='LE', data=filtered_df)
    plt.title(f"Graph of {country}'s LE over the years 2000-2015")
    ax.set_xticks(ax.get_xticks())  # Get the current tick positions
    ax.set_xticklabels(df['Year'].unique(), rotation=45, ha='right')
    ax.set_ylim(np.min(filtered_df["LE"])-1, np.max(filtered_df["LE"])+1)
    plt.xlabel("Year")
    plt.ylabel("Life expectancy")
    plt.xticks()
    plt.show()
    plt.clf()

for country in df["Country"].unique():
    graph(country)
```

```python
#Average life expectancy/GDP of each country? 
for country in df["Country"].unique():
    filtered_df = df[df["Country"] == f"{country}"]
    print(f"{country} Life Expectancy: " + str(np.mean(filtered_df["LE"])))
    print(f"{country} GDP: " + str(np.mean(filtered_df["GDP"])))
```
